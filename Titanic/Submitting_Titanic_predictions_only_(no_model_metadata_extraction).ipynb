##### **Train a model and generate predictions using X_test data**


```python
# Train a model and generate predictions using X_test data here
#Example prediction_labels for practice submission (replace with your own prediction list for real submissions!)

prediction_categories=['survived', 'died']

import random
predicted_labels=random.choices(prediction_categories, k=262)
```

#### **1. Submit Model predictions to leaderboard (without extracting model architecture information):**
- model metadata extraction allows you use compare_models() and instantiate_model() functions.

```python
#Submit Model predictions to leaderboard (without extracting model architecture information): 

#-- Generate predicted values (a list of predicted labels "survived" or "died") (Model 1)
prediction_labels = predicted_labels #model.predict(preprocessor(X_test))

# Submit Model 1 to Competition Leaderboard
mycompetition.submit_model(model_filepath = None,
                                 preprocessor_filepath=None,
                                 prediction_submission=prediction_labels)
```

##### Get the current leaderboard:

```python
data = mycompetition.get_leaderboard()
mycompetition.stylize_leaderboard(data)
```
